```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
df=pd.read_csv('cleaned_bike_data.csv')
```

```python
df.shape
```

```python
hourly_rides=df['hour'].value_counts().sort_index()
```

```python
plt.figure(figsize=(12,3))
plt.bar(hourly_rides.index, hourly_rides.values)
plt.title('Rides Per Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Rides')
plt.show()
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

daily_rides=df['day_of_week'].value_counts().reindex(day_order)
```

```python
plt.figure(figsize=(12,3))
plt.bar(daily_rides.index, daily_rides.values)
plt.title('Rides Per Day')
plt.xlabel('Day of Week')
plt.ylabel('Number of Rides')
plt.show()
```

```python
heatmap_data = pd.crosstab(df['day_of_week'], df['hour']).reindex(day_order)
plt.figure(figsize=(16, 6))
sns.heatmap(heatmap_data, cmap='YlOrRd')
plt.show()
```
