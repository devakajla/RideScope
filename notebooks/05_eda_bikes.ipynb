```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cleaned_bike_data.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
bike_counts=df['ride_type'].value_counts()
bike_counts
```

```python
plt.figure(figsize=(8,5))
plt.bar(bike_counts.index, bike_counts.values)
plt.title('Rides by Bike Type')
plt.ylabel('Number of Rides')
plt.show()
```

```python
avg_bike_duration=df.groupby('ride_type')['trip_duration'].mean()
avg_bike_duration
```

```python
plt.figure(figsize=(8,5))
plt.bar(avg_bike_duration.index, avg_bike_duration.values)
plt.title('Average Trip Duration by Bike Type')
plt.show()
```

```python
pd.crosstab(df['ride_type'], df['user_type']).plot(kind='bar', figsize=(8,5), title='Bike Type Usage: Member vs Casual')
plt.ylabel('Number of Rides')
plt.show()
```
