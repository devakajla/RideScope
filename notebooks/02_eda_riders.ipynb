```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cleaned_bike_data.csv')
```

```python
df.shape
```

```python
rider_counts=df['user_type'].value_counts()
rider_counts
```

Members: 183K ya 69%, Casuals: 81K ya 31%

```python
plt.figure(figsize=(8,5))
plt.bar(rider_counts.index, rider_counts.values)
plt.title('Totall Rides: Member vs Casual')
plt.ylabel('Number of Rides')
plt.show()
```

```python
avg_duration=df.groupby('user_type')['trip_duration'].mean()
avg_duration
```

**Average Durations:**

Casual --> 23.08 min

Member --> 11.85 min

```python
plt.figure(figsize=(8,5))
plt.bar(avg_duration.index, avg_duration.values)
plt.title('Average Ride Duration: Member vs Casual')
plt.ylabel('Average Time')
plt.show()
```

```python
pd.crosstab(df['hour'], df['user_type']).plot(kind='bar', figsize=(12,5), title='Rides Per Hour: Member vs Casual')
plt.ylabel('Number of Rides')
plt.show()
```

Peak hours are betweek 6-8 pm

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
pd.crosstab(df['day_of_week'], df['user_type']).reindex(day_order).plot(kind='bar', figsize=(10,5), title='Rides Per Day: Member vs Casual')
```

Most Rides are on Tuesdat and Wednesday
