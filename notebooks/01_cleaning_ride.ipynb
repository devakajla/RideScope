```python
from math import radians, sin, cos, sqrt, atan2
```

```python
def haversine(row):
    R = 6371   # Earth ka radius km mein (integer hai, 4 digit)

    lat1 = radians(row['start_lat'])
    lat2 = radians(row['end_lat'])        # end ka latitude column

    dlat = radians(row['end_lat'] - row['start_lat'])   # difference
    dlng = radians(row['end_lng'] - row['start_lng']) # difference

    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlng/2)**2

    return R * 2 * atan2(sqrt(a), sqrt(1 - a))
```

```python
df['distance_km'] = df.apply(haversine, axis=1)
```
