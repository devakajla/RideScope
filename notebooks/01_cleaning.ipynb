## Handling start_station_name and start_station_id null values


```python
start_nan=df[df['start_station_id'].isna()]
start_nan.shape
```

```python
start_nan[['start_lat','start_lng']].isna().sum()
```

```python
start_fill=df[df['start_station_id'].notna()]
```

```python
start_fill
```

```python
start_fill['start_station_name'].nunique()
```

```python
start_ref=start_fill.groupby('start_station_name')[['start_lat','start_lng']].median()
```

```python
start_ref
```

```python
from scipy.spatial import cKDTree
```

```python
start_ref.values
```

```python
tree=cKDTree(start_ref.values)
```

```python
tree
```

```python
dist, idx=tree.query(start_nan[['start_lat','start_lng']].values)
```

```python
start_ref.index[idx]
```

```python
df.loc[df['start_station_name'].isna(), 'start_station_name'] = start_ref.index[idx]
```

```python
df['start_station_name'].isna().sum()
```

```python
df['start_station_name']
```

```python
# Step 1: Mapping banao — name se id
name_to_id = df.dropna(subset=['start_station_id']).drop_duplicates('start_station_name').set_index('start_station_name')['start_station_id']

# Step 2: Null IDs fill karo using mapping
df.loc[df['start_station_id'].isna(), 'start_station_id'] = df.loc[df['start_station_id'].isna(), 'start_station_name'].map(name_to_id)
```

```python
df.isna().sum()
```

## Handling null values of end_station_name and end_station_id

```python
df=df.dropna(subset=['end_lat'])
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
end_fill=df[df['end_station_name'].notna()]
```

```python
end_fill.isna().sum()
```

```python
end_ref=df.groupby('end_station_name')[['end_lat','end_lng']].median()
```

```python
end_ref
```

```python
end_tree=cKDTree(end_ref.values)
```

```python
end_nan=df[df['end_station_name'].isna()]
```

```python
dist2, idx2 = end_tree.query(end_nan[['end_lat', 'end_lng']].values)
```

```python
df.loc[df['end_station_name'].isna(), 'end_station_name'] = end_ref.index[idx2]
```

```python
end_name_to_id = df.dropna(subset=['end_station_id']).drop_duplicates('end_station_name').set_index('end_station_name')['end_station_id']

df.loc[df['end_station_id'].isna(), 'end_station_id'] = df.loc[df['end_station_id'].isna(), 'end_station_name'].map(end_name_to_id)
```
